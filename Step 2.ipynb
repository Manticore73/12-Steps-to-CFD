```python
import numpy                        #importing numpy library
from matplotlib import pyplot       #2D plotting library
%matplotlib inline                 

nx = 50
dx = 2 / (nx - 1)
nt = 5                             #Total timesteps for which the sims is run
dt = 0.01                          #Duration for one timestep

u = numpy.ones(nx)                        #initializing the array 
u[int(0.5 / dx) : int((1/dx) + 1)] = 2    #Assigning '2' to certain members in the array
pyplot.plot(numpy.linspace(0,2,nx),u)     #Printing the initial condition 
```

```python
un = numpy.ones(nx)                 #Defining a new ARRAY 'un'
for n in range(nt):            
    un = u.copy()                   #Making it same as 'u'
    for i in range(1,nx):
        u[i] = un[i] - un[i]*(dt/dx)*(un[i] - un[i-1])     #Taylor series approximation of the DE
        pyplot.plot(numpy.linspace(0,2,nx),u)              #Plot that shows the convergence
        
```
