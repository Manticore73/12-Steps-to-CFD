```python
import numpy
from matplotlib import pyplot
%matplotlib inline

nx = 41                                      #Accept nx from user
dx = 2 / (nx - 1)  
nt = 20                                      #Dividing the grid into finite points
nu = 0.3                                     #Coefficeint of viscosity
sigma = 0.2                                  #Courants number
dt = sigma * dx**2 / nu                      #Timestep acceptable according to courants number

u = numpy.ones(nx)
u[int(0.5 / dx):int(1 / dx + 1)] = 2

un = numpy.ones(nx)

for n in range(nx):                          #Iterate through time
    un = u.copy()
    for i in range(1,nx-1):
        u[i] = un[i] + nu * (dt / dx**2) * ( un[i-1] - 2 * un[i] + un[i+1] )

pyplot.plot(numpy.linspace(0,2,nx),u);       
```
